### 1st Approach: Elbow Method 

```python
df1=df_cluster
```

```python
df1.head(2)
```

```python
# Scaling the new data frame for clustering
sc = skp.StandardScaler()
data_scale = np.array(df_cluster[['Lat', 'Lng']])
scaled = sc.fit_transform(data_scale.astype(float))
scaled_df = pd.DataFrame(scaled, columns=['Lattitude', 'Longitude'])
display(scaled_df)
```

```python
###Decide n-cluster using Elbow Method
wcss=[] ##sum of squared distance between each point
k_range = range(1,200)
for i in k_range:
    kmeans = skc.KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(scaled_df)
    wcss.append(kmeans.inertia_)
fig, ax = plt.subplots(figsize=(18, 8), dpi=120)
plt.plot(k_range, wcss, marker='o')
# for i, value in enumerate(wcss):
#     ax.text(i+1.05, value-0.005, round(value,1), fontsize=12, fontweight='bold')
    
plt.xticks(k_range)
# plt.grid()
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.savefig('num_clust.png')
plt.show()
```
